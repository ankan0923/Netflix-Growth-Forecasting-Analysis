```python
#extract year, month added
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')
df['year_added']=df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
```

**HYPOTHESIS TESTING**

```python
#Q1: Do Movies have significantly different durations across ratings?
movie_groups = [
    group['duration_num'].values
    for _, group in movies.dropna(subset=['duration_num', 'rating']).groupby('rating')
]

f_test = stats.f_oneway(*movie_groups)
print(f_test)
# Use the prepared movie_groups for ANOVA (avoid undefined 'table')
f_test = stats.f_oneway(*movie_groups)
print(f_test)
```

```python
#Q2:Is the proportion of Movies and TV Shows equal?
counts = df['type'].value_counts()
print(counts)

observed = counts.values
expected = [observed.sum() / len(observed)] * len(observed)

res = stats.chisquare(f_obs=observed, f_exp=expected)
print(res)
```

```python
from scipy.stats import chi2_contingency
counts = df['type'].value_counts()
chi2, p, dof, expected = chi2_contingency(counts)
print("Chi-square Statistic:", chi2)
print("p-value:", p)
```

```python
#Q3.Has Netflix added more content after 2017?
counts_per_year = df['year_added'].value_counts()

# Split the years into two groups
before_years = counts_per_year[counts_per_year.index < 2017]
after_years = counts_per_year[counts_per_year.index >= 2017]

# Perform t-test on the groups of yearly totals
ttest = stats.ttest_ind(before_years, after_years)
print(ttest)
```
